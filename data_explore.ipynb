```python
# Pivot table for Country:
pivot_1 = pd.pivot_table(df,index='Country')
pivot_1
```

```python
# Pivot table for Segment x Country:
pivot_2 = pd.pivot_table(df,index=['Segment','Country'])
pivot_2
```

```python
# Pivot table for product x country:
pivot_3 = pd.pivot_table(df,index=['Product','Country'], aggfunc='sum')
pivot_3
```

```python
type(pivot_3)
```

```python
# Sort data but just show COGS column:
pivot_3 = pd.pivot_table(df,index=['Product','Country'], values='COGS', aggfunc='sum')
pivot_3
```

```python
# Bar chart for sum of Products by countries
pivot_3 = pd.pivot_table(df,index=['Product'], values='COGS', aggfunc='sum').plot(kind='bar', color='maroon')
pivot_3
```

```python
pivot_3 = pd.pivot_table(df,index=['Country','Discount Band'], values='Manufacturing Price', aggfunc='sum')
pivot_3
```

```python
pivot_3 = pd.pivot_table(df,index=['Country'], values='Profit', aggfunc='sum').plot(kind='bar')
pivot_3
```
